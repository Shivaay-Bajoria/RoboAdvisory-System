```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

n_samples = 100000

#Numerical Values
age = np.random.randint(18, 65, n_samples)
income = np.random.randint(100000, 10000000, n_samples)
dependents = np.random.randint(0, 5, n_samples)
assets = np.random.randint(0, 5, n_samples)
savings_ratio = np.random.uniform(0.05, 0.5, n_samples)
debt_ratio = np.random.uniform(0.0, 0.6, n_samples)

#Categorical Values
financial_knowledge = np.random.choice(['Low', 'Medium', 'High'], n_samples, p=[0.3, 0.4, 0.3])
investment_goal = np.random.choice(['Wealth Creation', 'Retirement', 'Education', 'Short-term Savings'], n_samples)
```

```python
def assign_risk(age, income, dependents, assets, savings_ratio, debt_ratio, knowledge, goal):
    score = 0

    #Age's value on risk score
    if age < 35:
        score += 2
    elif age < 50:
        score += 1
    else:
        score -= 1

    #Income's value on risk score
    if income < 500000:
        score += 1
    elif income < 3000000:
        score += 2
    elif income < 7500000:
        score += 3
    else:
        score += 4

    #Dependent's value
    if dependents < 1:
        score += 2
    elif dependents < 3:
        score += 1
    else:
        score -= 1

    #Asset's Value
    if assets > 3:
        score += 2
    elif assets > 1:
        score += 1
    else:
        score -= 1

    #Saving_Ratio's Value
    if savings_ratio > 0.3:
        score += 2
    elif savings_ratio > 0.2:
        score += 1
    else:
        score -= 1

    #Debt_ratio's Value
    if debt_ratio > 0.4:
        score -= 2
    elif debt_ratio > 0.3:
        score -= 1

    #Knowledge's Value
    if knowledge == "High":
        score += 2
    elif knowledge == "Medium":
        score += 1
    else:
        score -= 1

    #Goal's Value
    if goal == 'Wealth Creation':
        score += 2
    elif goal == 'Retirement':
        score += 1
    elif goal == "Education":
        score += 1
    elif goal == 'Short-term Savings':
        score -= 1

    #Final Term
    if score >= 5:
        return "Aggresive"
    elif score >= 2:
        return "Balanced"
    else:
        return "Conservative"
```

```python
risk_level = [
    assign_risk(age[i], income[i], dependents[i], assets[i], savings_ratio[i], debt_ratio[i], financial_knowledge[i], investment_goal[i])
    for i in range(n_samples)
]

data = pd.DataFrame({
    'age': age,
    'income': income,
    'dependents': dependents,
    'assets': assets,
    'savings_ratio': savings_ratio,
    'debt_ratio': debt_ratio,
    'financial_knowledge': financial_knowledge,
    'investment_goal': investment_goal,
    'risk_level': risk_level
})

data.head(10)
```

```python
data.to_csv("created_data.csv", index=False)
print("Data Saved Succesfuly")
```
