```python
import pandas as pd
import os
```

```python
def add_features(input_path: str, output_path:str):
    df = pd.read_csv(input_path)

    #Mapping Categorical Data
    if "financial_knowledge" in df.columns:
        mapping = {"Low": 1, "Medium": 2, "High": 3}
        df["knowledge_score"] = df["financial_knowledge"].map(mapping)

    if "investment_goal" in df.columns:
        mapping = {
            "Wealth Creation": 1,
            "Retirement": 2,
            "Education": 3,
            "Short-term Savings": 4
        }
        df["goal_score"] = df["investment_goal"].map(mapping)

    if "risk_level" in df.columns:
        mapping = {
            "Conservative": 1,
            "Balanced": 2,
            "Aggresive": 3
        }
        df["risk_score"] = df["risk_level"].map(mapping)

    df = df.drop(columns=["financial_knowledge", "investment_goal", "risk_level"], errors="ignore")

    df.to_csv(output_path, index=False)
    print("Saved")
    return df
```

```python
add_features("../data/cleaned_data.csv", "../data/cleaned_data.csv")
```
