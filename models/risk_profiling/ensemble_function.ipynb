```python
import pandas as pd
import os
import joblib
import numpy as np
from sklearn.metrics import accuracy_score
from src.risk_models.custom_models import SoftMaxRegression
```

```python
X_test = pd.read_csv('../../data/X_test.csv')
y_test = pd.read_csv('../../data/y_test.csv').values.ravel()
```

```python
log_reg = joblib.load("../../models/risk_profiling/logisitic_model.joblib")
random_forest = joblib.load('../../models/risk_profiling/random_forest_model.joblib')
```

```python
proba_softmax = log_reg.predict_proba(X_test)
proba_rf = random_forest.predict_proba(X_test)
```

```python
avg_proba = (0.1 * proba_softmax) + (0.9 * proba_rf)
y_pred_ensemble = np.argmax(avg_proba, axis=1)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred_ensemble):.4f}")
```
