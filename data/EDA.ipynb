```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_data.CSV')
```

```python
continuous_features = ['age', 'income', 'assets', 'savings_ratio', 'debt_ratio']
df[continuous_features].hist(bins=30, figsize=(15, 10), layout=(2, 3))
plt.suptitle('Histograms of Continuous Features')
plt.show()
```

```python
plt.figure(figsize=(12, 9))
corr_matrix = df.corr()

# Use a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(ax=axes[0], x='goal_score', y='income', data=df)
axes[0].set_title('Income Distribution by Goal Score')

sns.boxplot(ax=axes[1], x='goal_score', y='age', data=df)
axes[1].set_title('Age Distribution by Goal Score')

plt.show()
```

```python
discrete_features = ['dependents', 'goal_score'] # Add your 'knowledge_' col name
fig, axes = plt.subplots(1, 2, figsize=(14, 5)) # Adjust '1, 2' based on col count

sns.countplot(ax=axes[0], x='dependents', data=df)
axes[0].set_title('Count of Dependents')

sns.countplot(ax=axes[1], x='goal_score', data=df)
axes[1].set_title('Count of Goal Scores')

plt.show()
```
